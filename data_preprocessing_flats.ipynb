```python
# removing links and property_id since they are not important
df.drop(columns = ['link','property_id'],inplace = True)
```

```python
# rename the "area" column as "price per sqft"
df.rename(columns={'area':'price_per_sqft'},inplace = True)
```

```python
# society is a categorical variable

print(df['society'].value_counts())
```

```python
# some societies have rating in the name itself , thus it creates an additional category,thus handling the society name
```

```python
df['society'].value_counts().shape  # total no. of categories = 636 , need to reduce it
```

```python
# keep only words in the name of the society
import re
df['society'] = df['society'].apply(lambda name:re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
# clean the price column
```

```python
df[df['price']=='Price on Request'].shape[0]

# there are 11 rows where price is not specified , thus we remove such rows

df = df[df['price']!='Price on Request']
```

```python
# the price column has 2 units lac and crore , making it a single unit "crore"

def price_conversion(x):
  if type(x)==float:
    return x
  else:
    if x[1] == 'Lac':
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)
df['price']  = df['price'].str.split(' ').apply(price_conversion)
```

## price /sq ft

```python
# clean the price per sqft column by removing rupee sign , comma and the unit

def clean_price_per_sqft(x):
  x = str(x)
  cleaned = re.sub(r'[^\d.]', '', x)
  cleaned = re.sub(r'\.+$', '', cleaned)
  if cleaned=='':
    return 0.00
  return float(cleaned)
df['price_per_sqft'] = df['price_per_sqft'].apply(clean_price_per_sqft)
```

## Bedroom , Bathrooms and balcony

```python
# handling bedrooms bathrooms and balcony

print(df['bedRoom'].isnull().sum())
print(df['bathroom'].isnull().sum())
print(df['balcony'].isnull().sum())
```

```python
print(df['bedRoom'].value_counts())
print(df['bathroom'].value_counts())
print(df['balcony'].value_counts())
```

```python
def clean_bedroom(x):
  return int(x[0])
df['bedRoom'] = df['bedRoom'].apply(clean_bedroom)

def clean_bathroom(x):
  return int(x[0])
df['bathroom'] = df['bathroom'].apply(clean_bathroom)

df['balcony']=df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

## Extra Rooms

```python
print(df['additionalRoom'].value_counts())
```

```python
print(df['additionalRoom'].isnull().sum())
```

```python
# there are 1305 flats with no information given on additional room , thus fill the null values with not available

df['additionalRoom']= df['additionalRoom'].fillna('not available')
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
print(df['additionalRoom'].value_counts())
```

## Floor num

```python
df.head()
```

```python
print(df['floorNum'].isnull().sum())
```

```python
print(df['floorNum'].value_counts())
```

```python
# extract the flat number and number of floors in the building
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

## Facing

```python
df['facing'].value_counts()
```

```python
print("null values : ",df['facing'].isnull().sum())
```

```python
df['facing']=df['facing'].fillna('NA')
```

```python
print(df['facing'].value_counts())
```

```python
df.insert(loc=4,column='area',value=round((df['price']*10000000)/df['price_per_sqft']))
df.insert(loc=1,column='property_type',value='flat')
```
